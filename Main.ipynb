# Data Cleaning


```python
# filling missing values in age column with mean
mean_age=round(   train["Age"].mean())



mean_age

train["Age"].fillna(mean_age, inplace=True)
```

```python
# convert the datatype of age column to int

train["Age"] = train["Age"].astype(int)
```

```python
train.info()
```

```python
# To convert a series to sting
#extracting the first letter of the cabins

train["Cabin"]
train["Cabin"].str[0]
train["Cabin"]=train["Cabin"].str[0]
```

```python
# Fill missing value in cabin colum with mode

mode_cabin = train["Cabin"].mode()[0]
train["Cabin"].fillna(mode_cabin, inplace=True)
```

```python
train["Cabin"]
```

```python
# Fill missing value in cabin colum with mode

mode_embarked = train["Embarked"].mode()[0]
train["Embarked"].fillna(mode_embarked, inplace=True)
```

```python
train["Embarked"]
```

# Data Exploration

This is the process of analysing the data to bring out hidden pattern in the datasets
- Uni-variate Analysis: analysing one colums at a time
- Bi-variate Analysis: analysin two columns at a time
- Multi-variate Analysis: analysing more than two columns at a time

```python
# Uni-variate Analysis
```

```python
survived=train["Survived"].value_counts()
survived.index=["dead","survived"]
survived
```

```python
# Pandas plot

survived.plot(kind="bar", color= ["orange", "green"])
plt.title("Count of Survival", color="blue")
plt.xticks(rotation=50, color="black")
plt.yticks(rotation=50, color="black")
plt.show()
```

```python
sex=train["Sex"].value_counts()
sex
```
